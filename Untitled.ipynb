```python
X = train.drop(['default_payment_next_month'], axis=1)
y = train.default_payment_next_month
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

#### Defining continuous and categorical features in dataset

```python
cont_features = ['limit_bal', 'age', 'bill_amt1', 'bill_amt2', 'bill_amt3',
                 'bill_amt4', 'bill_amt5', 'bill_amt6', 'pay_amt1', 'pay_amt2',
                'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6']

cat_features = ['sex', 'education', 'marriage', 'pay_0', 'pay_2',
               'pay_3', 'pay_4', 'pay_5', 'pay_6']
```

#### Assigning continuous and categorical features to pandas series and casting continuous values as floats for the standard scaler

```python
X_train_cont = X_train[cont_features].astype(float)
X_train_cat = X_train[cat_features]
X_test_cont = X_test[cont_features].astype(float)
X_test_cat = X_test[cat_features]
```

#### Defining index for train and test splits

```python
X_train_index = X_train.index
X_test_index = X_test.index
```

#### Scaling continuous features as dataframes

```python
ss = StandardScaler()
X_train_sc = pd.DataFrame(ss.fit_transform(X_train_cont), columns=X_train_cont.columns, index=X_train_index)
X_test_sc = pd.DataFrame(ss.transform(X_test_cont), columns=X_test_cont.columns, index=X_test_index)
```

#### One Hot Encoding categorical features

```python
ohe = OneHotEncoder(drop='if_binary')
X_train_coded = ohe.fit_transform(X_train_cat)
X_test_coded = ohe.transform(X_test_cat)
```

#### Retrieving feature names and casting encoded series to dataframes

```python
train_columns = ohe.get_feature_names(input_features=X_train_cat.columns)
test_columns = ohe.get_feature_names(input_features=X_test_cat.columns)
X_train_processed = pd.DataFrame(X_train_coded.todense(), columns=train_columns, index=X_train_index)
X_test_processed = pd.DataFrame(X_test_coded.todense(), columns=test_columns, index=X_test_index)
```

#### Concatenating scaled and encoded dataframes for entire processed train and test splits

```python
X_train_all = pd.concat([X_train_sc, X_train_processed], axis=1)
X_test_all = pd.concat([X_test_sc, X_test_processed], axis=1)
```

```python
X_train_all
```

```python
def evaluation(y, y_hat, title = 'Confusion Matrix'):
    '''takes in true values and predicted values.
    The function then prints out a classifcation report
    as well as a confusion matrix using seaborn's heatmap.'''
    cm = confusion_matrix(y, y_hat)
    precision = precision_score(y, y_hat, average = 'weighted')
    recall = recall_score(y, y_hat, average = 'weighted')
    f1 = f1_score(y, y_hat, average='weighted')
    accuracy = accuracy_score(y,y_hat)
    print(classification_report(y, y_hat))
    print('f1: ', f1)
    sns.heatmap(cm,  cmap= 'Greens', annot=True)
    plt.xlabel('predicted')
    plt.ylabel('actual')
    plt.title(title)
    plt.show()
```

#### It looks like we are running into a class imbalance problem, I am going to try oversampling the minority class with SMOTE

```python
train_ohe = train[cat_features]
train_cont = train[cont_features].astype(float)
train_index = train.index

ohe2 = OneHotEncoder(drop='if_binary')
ss2 = StandardScaler()
train_encoded = ohe2.fit_transform(train_ohe)
train_scaled = pd.DataFrame(ss2.fit_transform(train_cont), columns=train[cont_features].columns, index=train_index)
train_columns = ohe2.get_feature_names(input_features=train_ohe.columns)
train_processed = pd.DataFrame(train_encoded.todense(), columns=train_columns, index=train_index)
train_all = pd.concat([train_scaled, train_processed], axis=1)
```

```python
train_all['default_payment_next_month'] = train['default_payment_next_month']
```

```python
X_smote = train_all.drop(['default_payment_next_month'], axis=1)
y_smote = train_all.default_payment_next_month
smote = SMOTE(random_state=42, sampling_strategy='minority')
X_res, y_res = smote.fit_sample(X_smote, y_smote)
```
